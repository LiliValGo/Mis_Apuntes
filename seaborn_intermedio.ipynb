## Graficos categoricos

### Seaborn agrupa las graficas categoricas en tres grupos

### 1. Mostrar cada observación de forma individual 
### * stripplot().las observaciones se suporponen. dificil para ver datos individuales. usar jitter
### * swarmplot(): las observaciones no se superponen. no escala bien en grandes conjuntos de datos

### 2. Representaciones abstractas 
### * boxplot(): podemos ver varias medidas abstractas(media, max, min, cuartiles y valores atipicos)
### * violinplot(): combina kdeplot() y boxplot(). util para mostrar  grandes cantidades de datos 
### * boxenplot()

### 3. Estimaciones estadisticas 
### * barplot(): Muestra una estimación del valor e intervalos de confianza 
### * pointplot(): Es similar a barplot. Muestra una medida de resumen y un intervalo de confianza 
### * countplot()


## Mostrar cada observación de forma individual

```python
# stripplot
sns.stripplot(data=df_schoolimprovement2010,
         x='Award_Amount',
         y='Model Selected',
         jitter=True)

plt.show()
```

```python
# swarmplot
sns.swarmplot(data=df_schoolimprovement2010,
         x='Award_Amount',
         y='Model Selected',
         hue='Region')

plt.show()
```

## Representaciones abstractas

```python
# Create a boxplot
sns.boxplot(data=df_schoolimprovement2010,
         x='Award_Amount',
         y='Model Selected')

plt.show()
plt.clf()
```

```python
# Create a violinplot with the husl palette
sns.violinplot(data=df_schoolimprovement2010,
         x='Award_Amount',
         y='Model Selected',
         palette='husl')

plt.show()
plt.clf()
```

```python
# Create a lvplot with the Paired palette and the Region column as the hue
sns.boxenplot(data=df_schoolimprovement2010,
         x='Award_Amount',
         y='Model Selected',
         palette='Paired',
         hue='Region')

plt.show()
plt.clf()
```

## Estimaciones estadisticas 

```python
# Show a countplot with the number of models used with each region a different color
sns.countplot(data=df_schoolimprovement2010,
         y="Model Selected",
         hue="Region")

plt.show()
plt.clf()
```

```python
# Create a pointplot and include the capsize in order to show caps on the error bars
sns.pointplot(data=df_schoolimprovement2010,
         y='Award_Amount',
         x='Model Selected',
         capsize=.1)

plt.show()
plt.clf()
```

```python
# Create a barplot with each Region shown as a different color
sns.barplot(data=df_schoolimprovement2010,
         y='Award_Amount',
         x='Model Selected',
         hue='Region')

plt.show()
plt.clf()
```

## Gráficos de Matriz 

### La Grafica mas representativa de este grupo es el mapa de calor (heatmap()). Los graficos de matriz permite visualizar rapidamente las tendencias en un conjunto de datos.

### La funcion heatmap() de seaborn espera que los datos esten en una matriz.

### tambien podemos hacer uso de crosstab() para crear una tabla para resumir los datos. 

```python
# crosstab() para resumir datos 

df_crosstab= pd.crosstab(df_bike['mnth'], df_bike['weekday'], 
            values= df_bike['total_rentals'], 
            aggfunc = 'mean').round(0)         # con aggfunc podemos indicar la medida estadistica  
```

```python
df_crosstab
```

### Una vez tengamos esta tabla podemos construir el heatmap()

```python
sns.heatmap(pd.crosstab(df_bike['mnth'], df_bike['weekday'], 
            values= df_bike['total_rentals'], 
            aggfunc = 'mean'))
```

### En este tipo de grafico, los colores se aclaran a medida que aumentan los valores 

### personalizando el heatmap()

```python
sns.heatmap(df_crosstab, annot=True, fmt="g",     
           cmap='YlGnBu', cbar=False, linewidths=.5)

"""El fmt segun documentación y ejercicio de Datacamp debe ser igual a un str 'd'. 
Sin embargo, por digamos asi, jugar puse la letra 'g' y funciono """

"""con cbar=False, la barra de color no se mostrará"""

"""linewidrhs= ancho de linea"""
```

### centrando a heatmap()

```python
sns.heatmap(df_crosstab, annot=True, fmt="g",     
           cmap='YlGnBu', cbar=True, 
           center=df_crosstab.loc[9,6]) # seleccion del mes de septiembre(9) y el dia sabado(6)
```

## Usando joinGrid() y joinplot()


### JoinGrid() permite comparar la distribución de datos entre dos variables en dos pasos(definir el grid y asignar el tipo de gráfico). Hace uso de diagramas de dispersión, lineas de regresion, histogramas, diagramas de distribución y kde(estimaciones de densidad del nucleo). Veamos un ejemplo

```python
g = sns.JointGrid(data=df_collage, x="Tuition", y="ADM_RATE_ALL")
g.plot(sns.scatterplot, sns.histplot)
```

```python
g = sns.JointGrid(data=df_collage, x="Tuition", y="ADM_RATE_ALL")
g.plot_joint(sns.scatterplot, s=100, alpha=.5)
g.plot_marginals(sns.histplot, kde=True)
```

```python
g = sns.JointGrid(data=df_collage, x="Tuition", y="ADM_RATE_ALL")
g = g.plot_joint(sns.kdeplot)
g = g.plot_marginals(sns.kdeplot, shade=True)
g = g.annotate(stats.pearsonr)
```
