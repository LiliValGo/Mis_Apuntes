### Pero... ¿que es un objeto generador?

#### Un generador es como una lista de cpmprension excepto que no almacena la lista en la memoria.
#### No construye la lista, pero es un objeto sobre el cual se puede iterar para producir elementos de la lista segun sea necesario


```python
result = (num for num in range(6))
for num in result:
    print(num)
```

```python
result = (num for num in range(6))
print(list(result))                      # le pasamos un generador para crear una lista
```

```python
result = (num for num in range(6))
print(next(result))                      # le pasamos un generador next para iterar sus elementos 
```

### Funcion generadora 

#### La diferencia entre una funcion cualquiera y una funcion generadora es que al devolver valores no se hara con return sino con yield


```python
# funcion generadora

def num_sequence(n):
    """Gerera valores de 0 a n"""
    i = 0
    while i < n:
        yield i
        i += 1
        
result = num_sequence(5)
print(type(result))
```

```python
for item in result:    # Ahora podemos iterar sobre la funcion con un bucle 
    print(item)
```

## Ejercicio 2. 

### Se le da una lista de cadenas lannister y, usando una expresión generadora, crea un objeto generador sobre el que iterará para imprimir la longitud de sus valores.

```python
# Create a list of strings: lannister
lannister = ['cersei', 'jaime', 'tywin', 'tyrion', 'joffrey']

# Create a generator object: lengths
lengths = (len(person) for person in lannister)

# Iterate over and print the values in lengths
for value in lengths:
    print(value)
```

## Ejercicio 3. 

### creará una función generadora con un mecanismo similar al de la expresión generadora que definió en el ejercicio anterior:

```python
# Create a list of strings
lannister = ['cersei', 'jaime', 'tywin', 'tyrion', 'joffrey']

# Define generator function get_lengths
def get_lengths(input_list):
    """Generator function that yields the
    length of the strings in input_list."""

    # Yield the length of a string
    for person in input_list:
        yield len(person)

# Print the values generated by get_lengths()
for value in get_lengths(lannister):
    print(value)
```

### utilizará una lista de comprensión para extraer la hora de los datos de Twitter con marca de tiempo. Agregará una expresión condicional a la comprensión de la lista para que solo seleccione las horas en las que la entrada [17:19] es igual a '19'
