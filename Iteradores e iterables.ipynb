# Introduction to iterators

### Iterando sobre un bucle for 

```python
employee = ['Nick', 'Lore', 'Hugo']
```

```python
for employee in employees:
    print(employee)
```

```python
for letter in 'Datacamp':
    print(letter)
```

### Utlizando la función iter()

```python
word = 'Da'
it = iter(word)
next(it)
```

```python
next(it)
```

```python
next() # Cuando el iterador ha entregado su ultimo elemento y solicitamos el prómo elemento - next(), este nos dara typeError 
```

```python
# Al igual que con bucle for. Tambien podemos imprimir la totalidad de los valores.

word = 'Data'
it = iter(word)
print(*it)           # El caracter * se denomina operador splat 
```

### Iterando sobre diccionarios

```python
pythonistas = {'Hugo': 'Bowne-Anderson', 'Francis': 'Castro'}

for key, value in pythonistas.items():
    print(key, value)
    
```

# Playing with iterators

En este capitulo abordaremos dos funciones importantes para manejar el tema de los iteradores:
1. enumerate()  

2. zip()


```python
# enumerate()

avengers = ['howkeye', 'iron man', 'thor', 'quicksilver']

e = enumerate(avengers)  # con start = n. podemos modificar el valor de enumerate()
print(type(e))
```

```python
e_list = list(e)   # Convertimos el objeto enumerate() en una lista para visualizar su contenido

print(e_list)
```

De esta forma se estructura en bucle 

```python
avengers = ['howkeye', 'iron man', 'thor', 'quicksilver']

for index, value in enumerate(avengers, start = 20):  # con start = 20. podemos modificar el valor de enumerate()
    print(index, value)
```

```python
# zip()

avengers = ['howkeye', 'iron man', 'thor', 'quicksilver']
names = ['borton', 'stark', 'odinson', 'maximoff']

z = zip(avengers, names)

print(type(z))
```

```python
z_list = list(z)

print(z_list)  # Convertimos el objeto zip() en una lista para visualizar su contenido
```

```python
De esta forma se estructura en bucle 
```

```python
for z1, z2 in zip(avengers, names):
    print(z1, z2)
```

De esta forma se imprime todos los elementos del zip()

```python
z = zip(avengers, names)
print(*z)      # El caracter * se denomina operador splat 
```
